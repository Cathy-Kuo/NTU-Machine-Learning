```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('train.csv', index_col=0)
df_train = df_train[df_train['Churn Category'].notnull()]
train_end_idx = len(df_train) 
df_test = pd.read_csv('test.csv', index_col=0)
#score:1 (Very Unsatisfied) to 5 (Very Satisfied)
#score=4 or 5 -> Churn Category=0.0 or nan

# 為了處理方便，把 'train.csv' 和 'test.csv' 合併起來，'test.csv'的 RainToday 欄位用 0 補起來。
# 以 train_end_idx 作為 'train.csv' 和 'test.csv' 分界列，
df = pd.concat([df_train, df_test], sort=False)
df.drop(columns=['Customer ID'], inplace=True)
```

```python
#CV GridSearch
import xgboost as xgb
from sklearn.metrics import accuracy_score, f1_score

X_train, y_train = df.values[:train_end_idx, :], df_churn.values[:train_end_idx]
params = {}
params['objective'] = 'multi:softprob'
params['eval_metric'] = 'mlogloss'
params['num_class'] = 6
params['learning_rate'] = 0.01

gridsearch_params = [
    (max_depth, min_child_weight, eta, obj)
    for max_depth in range(5,12)
    for min_child_weight in range(2,6)
    for eta in [0.01, 0.05, 0.1, 0.15, 0.2]
]

# Define initial best params and MAE
min_mae = float("Inf")
best_params = None
time = 0
for max_depth, min_child_weight, eta, obj in gridsearch_params:
    print(time)
    time+=1
    print("CV with max_depth={}, min_child_weight={}, eta={}".format(max_depth,min_child_weight, eta))
    # Update our parameters
    params['max_depth'] = max_depth
    params['min_child_weight'] = min_child_weight
    params['eta'] = eta
    d_train = xgb.DMatrix(X_train, label=y_train)

    
    # Run CV
    cv_results = xgb.cv(
        params,
        d_train,
        num_boost_round=999,
        seed=42,
        nfold=5,
        early_stopping_rounds=100
    )
    # Update best MAE
    mean_mae = cv_results['test-mlogloss-mean'].min()
    boost_rounds = cv_results['test-mlogloss-mean'].argmin()
    print("\tMAE {} for {} rounds".format(mean_mae, boost_rounds))
    if mean_mae < min_mae:
        min_mae = mean_mae
        best_params = (max_depth,min_child_weight,eta, obj)
print("Best params: {}, {},{}, mlogloss: {}".format(best_params[0], best_params[1],best_params[2], min_mae))

# get the best score and update params
# params['max_depth'] = 10
# params['min_child_weight'] = 6
```

```python
def satis_45_label_to_0(df_test, y_test):
    satisfaction = list(df_test['Satisfaction Score'])
    for i, score in enumerate(satisfaction):
        if score==4.0 or score==5.0:
            y_test[i] = 0
    return y_test
```

```python
y_test_45 = satis_45_label_to_0(df_test, y_test)

df_ans_45 = pd.DataFrame({'Customer ID': df_test['Customer ID'], 'Churn Category':y_test_45})
df_ans_45.to_csv('df_ans_45.csv', index=False)
```
