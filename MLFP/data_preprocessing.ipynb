```python
import pandas as pd
df_demo = pd.read_csv('demographics.csv').fillna('nan')
df_loc = pd.read_csv('location.csv').fillna('nan')
df_pop = pd.read_csv('population.csv').fillna('nan')
df_satis = pd.read_csv('satisfaction.csv').fillna('nan')
df_serv = pd.read_csv('services.csv').fillna('nan')
df_stat = pd.read_csv('status.csv').fillna('nan')
df_test = pd.read_csv('Test_IDs.csv')
df_train = pd.read_csv('Train_IDs.csv')
```

```python
#'Under 30'&'Senior Citizen' similar to age, 'Dependents' similar to 'Number of Dependents'
demo = df_demo.drop(columns=['Count','Under 30','Senior Citizen','Dependents']).set_index('Customer ID').T.to_dict()
```

```python
train_id = list(df_train['Customer ID'])
test_id = list(df_test['Customer ID'])
```

```python
Gender = []
Age = []
Married = []
Number_Dependents = []
for idx in train_id:
    if idx in demo:
        Gender.append(demo[idx]['Gender'])
        Age.append(demo[idx]['Age'])
        Married.append(demo[idx]['Married'])
        Number_Dependents.append(demo[idx]['Number of Dependents'])
    else:
        Gender.append('nan')
        Age.append('nan')
        Married.append('nan')
        Number_Dependents.append('nan')

df_train['Gender'] = Gender
df_train['Age'] = Age
df_train['Married'] = Married
df_train['Number of Dependents'] = Number_Dependents
df_train
```
